```python
import pandas as pd
from MyUtil import OpenCv_Util
import os

# 定义一个函数来合并文件夹下的所有Excel文件
def merge_excel_files(folder_path):
    # 初始化一个空的DataFrame用于存储合并后的数据
    merged_df = pd.DataFrame()
    # 遍历文件夹中的所有文件
    for file in os.listdir(folder_path):
        # 检查文件扩展名是否为.xlsx（Excel文件）
        if file.endswith('.xlsx'):
            file_path = os.path.join(folder_path, file)  # 获取文件完整路径
            df = pd.read_excel(file_path)  # 读取Excel文件
            merged_df = pd.concat([merged_df, df], ignore_index=True)  # 合并数据
    return merged_df
df_job_original_dir = f"./original_output_compare/job_original"
df_fail_url_dir =  f"./original_output_compare/fail_url"
df_error_url_dir = f"./original_output_compare/error_url"
df_company_original_dir =  f"./original_output_compare/company_original"
df_job_data_compare_dir =  f"./original_output_compare/job_data_compare"
df_company_data_compare_dir =  f"./original_output_compare/company_data_compare"
df_job_list_data_compare_dir =  f"./original_output_compare/job_list_data_compare"

df_dirs = []
df_dirs.append(df_job_original_dir)
df_dirs.append(df_fail_url_dir)
df_dirs.append(df_error_url_dir)
df_dirs.append(df_company_original_dir)
df_dirs.append(df_job_data_compare_dir)
df_dirs.append(df_company_data_compare_dir)
df_dirs.append(df_job_list_data_compare_dir)
# 创建一个新文件夹用于存储合并后的Excel文件
OpenCv_Util.mkdir('./original_output_compare/merge')
for df_dir in df_dirs:
    data = merge_excel_files(df_dir)
    df_dir = df_dir.replace('./original_output_compare/','')
    data.to_excel(f'./original_output_compare/merge/{df_dir}_merged.xlsx', index=False)
    print(df_dir,'处理完毕')
```
