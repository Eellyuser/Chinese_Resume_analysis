## this code is the main function,the previous code 'if 1' is used to hide code which is useless,it can be stoped anytime and then please save variable

```python
if 1:
    for index, data in segment_data.iterrows():
        url = data['link']
        edge_driver.get_driver(url)
        id = data['id']
        company_name = data['company']
        info = {
            "url":url,
            "id":id,
            "company_name":company_name,
            "error_info":""
        }
        def collect_error():
            if edge_driver.get_element_text('//*[@id="main"]/div/div/div/a') == "首页":
                info['error_info'] = f"第{id}号{company_name}链接失效"
                print(info['error_info'])
                error_url.append(info)
                return True
            elif not edge_driver.wait_element(xpath='//*[@id="header"]/div[1]/div[2]/ul/li[1]/a',timeout=6,checktext="首页"):
                info['error_info'] = f"第{id}号{company_name}加载失败"
                print(info['error_info'])
                fail_url.append(info)
                return True
            elif edge_driver.get_element_attribute('//*[@id="wrap"]/div[3]/div/div[1]/div[1]/form/div[2]/p/input',attribute='placeholder') == "搜索职位、公司":
                info['error_info'] = f"第{id}号{company_name}链接失效超时返回主页"
                print(info['error_info'])
                error_url.append(info)
                return True
            else:
                return False
        def get_job():
            # 职位相关字段
            job_city = job_experience = job_degree = job_tags = job_keyword = ""
            job_city = edge_driver.get_element_text("text-desc.text-city",By.CLASS_NAME)
            job_experience = edge_driver.get_element_text("text-desc.text-experiece",By.CLASS_NAME)
            job_degree = edge_driver.get_element_text("text-desc.text-degree",By.CLASS_NAME)
            job_tags = edge_driver.get_elements_text('//*[@id="main"]/div[1]/div/div/div[2]/div[2]/span')
            if job_tags != "Null":
                job_tags = ",".join(i for i in job_tags)
            job_tags
            job_keyword = edge_driver.get_elements_text('//*[@id="main"]/div[3]/div/div[2]/div[1]/ul/li')
            if job_keyword != "Null":
                job_keyword = ",".join(i for i in job_keyword)
            job_keyword
            job_data = {
                "id":id,
                "company_name":company_name,
                "job_city":job_city,
                "job_experience":job_experience,
                "job_degree":job_degree,
                "job_tags":job_tags,
                "job_keyword":job_keyword
            }
            pprint(job_data)
            job.append(job_data)
            print(f"{company_name}职业信息已爬取完毕")
        def get_company():
            # 公司相关字段
            # 公司简介
            abbreviate = abbr_html = abbr_job = stage = scale = industry = ""
            abbreviate_father = edge_driver.get_element(xpath='sider-company',xpath_kind=By.CLASS_NAME)
            if abbreviate_father != "Null":
                abbreviate = edge_driver.get_element_text(father_element=abbreviate_father,xpath='./div[1]/a[2]')
                # 公司网址
                abbr_html = edge_driver.get_element_attribute(father_element=abbreviate_father, xpath='./div[1]/a[2]',attribute='href')
                # 职业网址
                abbr_job = edge_driver.get_element_attribute(father_element=abbreviate_father,xpath_kind=By.CLASS_NAME,xpath='look-all',attribute='href')
                # 企业阶段（是否融资）
                # stage_son = edge_driver.get_element(father_element=abbreviate_father, xpath_kind=By.CLASS_NAME, xpath='icon-stage')
                # stage = edge_driver.get_element_text(father_element=stage_son, xpath='./parent::p')
                stage = edge_driver.get_element_text(father_element=abbreviate_father, xpath="//p[.//i[contains(@class, 'icon-stage')]]")
                # 企业规模
                scale = edge_driver.get_element_text(father_element=abbreviate_father, xpath="//p[.//i[contains(@class, 'icon-scale')]]")
                # 企业种类
                industry = edge_driver.get_element_text(father_element=abbreviate_father, xpath="//p[.//i[contains(@class, 'icon-industry')]]")
            else:
                print("公司简介为空")
            # 公司工商信息
            # 点击事件
            edge_driver.click_element('look-all-btn',By.CLASS_NAME)
            job_detail = edge_driver.get_element('level-list',By.CLASS_NAME)
            company_user =  res_time = company_type = manager_state = company_fund = ""
            if job_detail != "Null":
                company_user  =  edge_driver.get_element_text("company-user",By.CLASS_NAME,job_detail).replace('法定代表人\n','')
                res_time      =  edge_driver.get_element_text("res-time",By.CLASS_NAME,job_detail).replace('成立日期\n','')
                company_type  =  edge_driver.get_element_text("company-type",By.CLASS_NAME,job_detail).replace('企业类型\n','')
                manager_state =  edge_driver.get_element_text("manage-state",By.CLASS_NAME,job_detail).replace('经营状态\n','')
                company_fund  =  edge_driver.get_element_text("company-fund",By.CLASS_NAME,job_detail).replace('注册资金\n','')
            else:
                print("公司工商信息为空")
            company_data={
                "id":id,
                "company_name":company_name,
                # 分界线
                "abbreviate":abbreviate,
                "abbr_html":abbr_html,
                "abbr_job":abbr_job,
                "stage":stage,
                "scale":scale,
                "industry":industry,
                # 分界线
                "company_user":company_user,
                "res_time":res_time,
                "company_type":company_type,
                "manager_state":manager_state,
                "company_fund":company_fund
            }
            pprint(company_data)
            company.append(company_data)
            print(f"{company_name}公司信息爬取完毕")
        def get_similar_job():
            similar_job_list = edge_driver.get_elements("//ul[@class='similar-job-list']/li")
            similar_job_html = similar_job_name = similar_job_salary = similar_job_company = similar_job_location = similar_job_compare = ""
            for i in similar_job_list:
                similar_job_html = edge_driver.get_element_attribute(xpath="./a[1]",father_element=i,attribute='href')
                similar_job_name = edge_driver.get_element_text(xpath="./a[1]/div[1]/span[1]",father_element=i)
                similar_job_salary = edge_driver.get_element_text(xpath="./a[1]/div[1]/span[2]",father_element=i)
                similar_job_company = edge_driver.get_element_text(xpath="./a[1]/p[1]/span[1]/span[2]",father_element=i)
                similar_job_location = edge_driver.get_element_text(xpath="./a[1]/p[1]/span[2]",father_element=i)
                similar_job_compare = edge_driver.get_element_attribute(xpath="./a[2]",father_element=i,attribute='href')
                similar_job_data = {
                    "id":id,
                    "company_name":company_name,
                    "similar_job_html":similar_job_html,
                    "similar_job_name":similar_job_name,
                    "similar_job_salary":similar_job_salary,
                    "similar_job_company":similar_job_company,
                    "similar_job_location":similar_job_location,
                    "similar_job_compare":similar_job_compare
                }
                pprint(similar_job_data)
                similar_job_list_data.append(similar_job_data)
                if similar_job_html == similar_job_name == similar_job_salary == similar_job_company == similar_job_location == similar_job_compare == "Null":
                    fail_url.append(info)
                    print("相似职位爬取失败！")
                    break
            print(f"{company_name}相似职位爬取完毕")
        def get_look_job():
            look_job_list_father = edge_driver.get_element("look-job-box",By.CLASS_NAME)
            look_job_list = edge_driver.get_elements("./ul/li",By.XPATH,look_job_list_father)
            look_job_html = look_job_name = look_job_salary = look_job_company = look_job_location = ""
            for i in look_job_list:
                look_job_html = edge_driver.get_element_attribute(xpath="./a",father_element=i,attribute='href')
                look_job_name = edge_driver.get_element_text(xpath="./a[1]/div[1]/div[1]",father_element=i)
                look_job_salary = edge_driver.get_element_text(xpath="./a[1]/div[1]/p[1]",father_element=i)
                look_job_company = edge_driver.get_element_text(xpath="./a[1]/div[2]/div[1]/span[2]",father_element=i)
                look_job_location = edge_driver.get_element_text(xpath="./a[1]/div[2]/div[2]",father_element=i)
                data = {
                    "id":id,
                    "company_name":company_name,
                    "look_job_html":look_job_html,
                    "look_job_name":look_job_name,
                    "look_job_salary":look_job_salary,
                    "look_job_company":look_job_company,
                    "look_job_location":look_job_location,
                }
                pprint(data)
                look_job_list_data.append(data)
                if look_job_html == look_job_name == look_job_salary == look_job_company == look_job_location == "Null":
                    fail_url.append(info)
                    print("还看了职位爬取失败!")
                    break
            print(f"{company_name}看过该职位的人还看了爬取完毕")
        def get_omnibus_job():
            omnibus_job_list_father = edge_driver.get_element("omnibus-slider-main",By.CLASS_NAME)
            omnibus_job_list_li = edge_driver.get_elements("./ul[1]/li[1]/ul[1]/li",By.XPATH,omnibus_job_list_father)
            for i in omnibus_job_list_li:
                omnibus_job_html = omnibus_job_name = omnibus_job_salary = omnibus_job_company = omnibus_job_location = ""
                omnibus_job_html = edge_driver.get_element_attribute("./a[1]",By.XPATH,i,attribute='href')
                omnibus_job_name = edge_driver.get_element_text("./a[1]/div[1]/div[1]",By.XPATH,i)
                omnibus_job_salary = edge_driver.get_element_text("./a[1]/div[1]/p",By.XPATH,i)
                omnibus_job_company =  edge_driver.get_element_text("./a[1]/div[2]/div[1]/span[2]",By.XPATH,i)
                omnibus_job_location = edge_driver.get_element_text("./a[1]/div[2]/div[2]",By.XPATH,i)
                omnibus_data = {
                    "id":id,
                    "company_name":company_name,
                    "omnibus_job_html":omnibus_job_html,
                    "omnibus_job_name":omnibus_job_name,
                    "omnibus_job_salary":omnibus_job_salary,
                    "omnibus_job_company":omnibus_job_company,
                    "omnibus_job_location":omnibus_job_location,
                }
                pprint(omnibus_data)
                omnibus_job_list_data.append(omnibus_data)
                if omnibus_job_html == omnibus_job_name == omnibus_job_salary == omnibus_job_company == omnibus_job_location == "Null":
                    fail_url.append(info)
                    print("精选职位爬取失败!")
                    break
            print(f"{company_name}精选职位爬取完毕")
        if collect_error():
            continue
        get_job()
        edge_driver.click_element('detail-hasmore-btn',By.CLASS_NAME)
        get_similar_job()
        get_look_job()
        get_omnibus_job()
        get_company()
        print(f"第{id}号{company_name}爬取完毕")
        true_end = true_end + 1
```

# this code is used to merge the data from the folder

```python
import pandas as pd
from MyUtil import OpenCv_Util
import os

# 定义一个函数来合并文件夹下的所有Excel文件
def merge_excel_files(folder_path):
    # 初始化一个空的DataFrame用于存储合并后的数据
    merged_df = pd.DataFrame()

    # 遍历文件夹中的所有文件
    for file in os.listdir(folder_path):
        # 检查文件扩展名是否为.xlsx（Excel文件）
        if file.endswith('.xlsx'):
            file_path = os.path.join(folder_path, file)  # 获取文件完整路径
            df = pd.read_excel(file_path)  # 读取Excel文件
            merged_df = pd.concat([merged_df, df], ignore_index=True)  # 合并数据
    return merged_df
df_look_job_list_data_dir = f"./output/look_job_list_data"
df_fail_url_dir =  f"./output/fail_url"
df_error_url_dir = f"./output/error_url"
df_omnibus_job_list_data_dir =  f"./output/omnibus_job_list_data"
df_similar_job_list_data_dir =  f"./output/similar_job_list_data"
df_company_dir =  f"./output/company"
df_job_dir =   f"./output/job"
df_dirs = []
df_dirs.append(df_look_job_list_data_dir)
df_dirs.append(df_fail_url_dir)
df_dirs.append(df_error_url_dir)
df_dirs.append(df_omnibus_job_list_data_dir)
df_dirs.append(df_similar_job_list_data_dir)
df_dirs.append(df_company_dir)
df_dirs.append(df_job_dir)
# 创建一个新文件夹用于存储合并后的Excel文件
OpenCv_Util.mkdir('./output/merge')
for df_dir in df_dirs:
    data = merge_excel_files(df_dir)
    df_dir = df_dir.replace('./output/','')
    data.to_excel(f'./output/merge/{df_dir}_merged.xlsx', index=False)
    print(df_dir,'处理完毕')
```
