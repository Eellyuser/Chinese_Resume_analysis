```python
import re
from Util.util import get_index_by_automaton,get_log_config
log_config = get_log_config()
log_info = "101"
def get_text_index(txt_text, search_string, tag):
    """
    Function for tagging fields in the txt_text.
    :param txt_text: The text where to search the string.
    :param search_string: The string to be searched, acting as the returned key value for easy searching.
    :param tag: The tag.
    :return: list
    """
    automaton_info = []
    # If the text is found
    if txt_text.find(search_string) != -1:
        start_idx = txt_text.find(search_string)
        end_idx = start_idx + len(search_string) - 1
        automaton_info = [{
            search_string: {
                "start_index": start_idx,
                "end_index": end_idx,
                "tag": tag,
            }
        }]
        log_config.info(f"{log_info} file {search_string} tagged as {tag}")
    else:
        pattern = r'[、，。；：？！,.;:?!]'
        # Split the search_string using the regex pattern
        segments = re.split(pattern, search_string)
        if segments:  # Ensure there's at least one segment to search
            for segment in segments:
                if segment and txt_text.find(segment) != -1:
                    start_idx = txt_text.find(segment)
                    end_idx = start_idx + len(segment) - 1
                    automaton_info.append({
                        segment: {
                            "start_index": start_idx,
                            "end_index": end_idx,
                            "tag": tag,
                        }
                    })
                    log_config.info(f"{log_info} file {segment} tagged as {tag}")
            if not automaton_info:  # If no segments found
                log_config.warning(f"{log_info} 文件 {search_string} 未能成功标注")
        else:
            log_config.error(f"{log_info} 文件 {search_string} 未找到任何可标注的段落")
    return automaton_info

def process_json_text(json_text, txt_text, key_mapping):
    all_tag_list = []
    for key, value in json_text.items():
        new_key = key_mapping.get(key, key)  # Use get to handle missing keys gracefully
        if isinstance(value, list):
            value = ", ".join(value)  # Convert list to string if needed
        tags = get_text_index(txt_text, value, new_key)
        for tag in tags:
            all_tag_list.append(tag)
    return all_tag_list
```
