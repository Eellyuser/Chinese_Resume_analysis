```python
def marking_tag(json_text,txt_text,key_mapping=key_mapping,log_info="101"):
    """
    打标记
    :param json_text: 原始标注好的json文件
    :param txt_text: 原始文本
    :param key_mapping: 简写映射
    :param log_info: 日志文本
    :return:
    """
    import re
    from Util.util import get_index_by_automaton,get_log_config
    from pprint import pprint
    log_config = get_log_config()
    def get_text_index(search_string,tag):
        """
        用于标注字段
        :param search_string:被查找的字符串,充当返回的key值，便于查找
        :param tag: 标注的标志
        :param name: logging信息，文件名称便于定位错误信息
        :return:list
        """
        # 变量初始化
        automaton_info = []
        # 若找得到文本
        if txt_text.find(search_string) != -1:
            start_idx = txt_text.find(search_string)
            end_idx = start_idx + len(search_string) - 1
            automaton_info = [{
                search_string: {
                    "start_index": start_idx,
                    "end_index": end_idx,
                    "tag": tag,
                }
            }]
            log_config.info(f"{log_info}文件 {search_string} 标注为 {tag}")
            # print(automaton_info)
        # 找不到文本
        else:
            pattern = r'[、，。；：？！,.;:?!]'
            # 使用正则表达式分割文本
            segments = re.split(pattern, search_string)
            segments.pop()
            # print(segments)
            log_config.warning(f"{log_info}文件 {search_string} 标注为 :")
            log_config.error(f"{log_info}文件 {search_string} 未标注为 :")
            automaton_info = get_index_by_automaton(segments,tag,txt_text)
            # pprint(automaton_info)
        return automaton_info
    all_tag_list = []
    # 遍历一级结构字典
    # pprint(json_text)
    for key, value in json_text.items():
        if isinstance(value, str):  # 如果字段值是字符串，则直接添加
            new_key = key_mapping[key]
            for split_tag in get_text_index(value,new_key):
                all_tag_list.append(split_tag)
        elif isinstance(value, list):
            # 遍历二级结构列表
            for (index,item) in enumerate(value):
                # 遍历字典
                for sub_key, sub_value in item.items():
                    new_key = key_mapping[key]+"_"+key_mapping[sub_key]+"_"+str(index)
                    split_tag_list = get_text_index(sub_value,new_key)
                    pprint(split_tag_list)
                    # print(index,item)
                    for split_tag in split_tag_list:
                        all_tag_list.append(split_tag)
    return all_tag_list
all_tag = marking_tag(ner_json,text)
from pprint import pprint
# pprint(all_tag)
```

```python
# 解码程序
def decode_tagged_text(tagged_text_lines):
    entities = []
    current_entity = []
    current_type = None

    for line in tagged_text_lines:
        parts = line.strip().split()  # 将行分割为字符和标注
        if len(parts) == 2:  # 确保行有两个部分
            char, tag = parts
        else:
            continue  # 如果行不符合预期格式，跳过这行

        if tag.startswith('B-'):
            if current_entity:
                entities.append((''.join(current_entity), current_type))
                current_entity = []
            current_type = tag.split('-')[1].split('|')[0]
            current_entity.append(char)
        elif tag.startswith('I-') and current_type:
            current_entity.append(char)
        elif tag.startswith('L-') and current_type:
            current_entity.append(char)
            entities.append((''.join(current_entity), current_type))
            current_entity = []
            current_type = None
        elif tag == 'O':
            if current_entity:
                entities.append((''.join(current_entity), current_type))
                current_entity = []
                current_type = None

    # 捕获最后一个实体，如果它在文本末尾
    if current_entity:
        entities.append((''.join(current_entity), current_type))
    return entities
# 文件路径
tagged_text_path = './Corpus_Path/2024年03月03日15时01分/101/decoding/tag_with_1.txt'
# 读取标注文本
try:
    with open(tagged_text_path, 'r', encoding='utf-8') as file:
        tagged_text_lines = file.readlines()
    # 使用改进的函数进行解码
    entities = decode_tagged_text(tagged_text_lines)
except FileNotFoundError:
    print(f"文件未找到: {tagged_text_path}")
except Exception as e:
    print(f"读取文件时发生错误: {e}")
result_path = './Corpus_Path/2024年03月03日15时01分/101/decoding/result_1.txt'
entities_str = '\n'.join([f"{entity} {entity_type}" for entity, entity_type in entities])
with open(result_path,'w',encoding='utf-8')as f:
    f.write(entities_str)
print(entities_str)
```
